```python
import os

def compare_folders(folder1_path, folder2_path):
    missing_files = []
    excess_files = []

    for root, _, files in os.walk(folder1_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            relative_path = os.path.relpath(file_path, folder1_path)
            missing_file_path = os.path.join(folder2_path, relative_path)
            if not os.path.exists(missing_file_path):
                missing_files.append(missing_file_path)

    for root, _, files in os.walk(folder2_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            relative_path = os.path.relpath(file_path, folder2_path)
            excess_file_path = os.path.join(folder1_path, relative_path)
            if not os.path.exists(excess_file_path):
                excess_files.append(excess_file_path)

    return missing_files, excess_files

folder1_path = "D:\\comic"
folder2_path = "G:\\comic"

missing_files, excess_files = compare_folders(folder1_path, folder2_path)

print("Missing files:")
for file in missing_files:
    print(file)

print("Excess files:")
for file in excess_files:
    print(file)
```
